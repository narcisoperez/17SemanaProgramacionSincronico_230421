## <u>Abstraccin </u>
### (no confundir abstracion con clases abstractas)
El concepto de <b>encapsulamiento</b> descripto debajo se apoya sobre el concepto de abstraccin.

Objects in an OOP language provide an <b>abstraction that hides the internal implementation details.</b> Similar to the coffee machine in your kitchen, you just need to know which methods of the object are available to call and which input parameters are needed to trigger a specific operation. But you dont need to understand how this method is implemented and which kinds of actions it has to perform to create the expected result.

Data Abstraction may also <b>be defined as the process of identifying only the required characteristics of an object ignoring the irrelevant details.</b> The properties and behaviors of an object differentiate it from other objects of similar type and also help in classifying/grouping the objects.

En POO solo necesita saber como interaccionar con los objetos, no necesita conocer los detalles de cmo est implementada la clase a partir de la cual se instancia el objeto. Slo necesita conocer su interfaz pblica.

La encapsulacin es una forma de abstraccin, ademas es un mecanismo para llevar a la prctica la abstraccin.

El nivel de abstraccin puede ser bajo (en un objeto se manipulan datos y mtodos individualmente), o alto (en un objeto solo se usan sus mtodos de servicio).

## <u>Clases abstractas</u>

Un concepto importante en programacin orientada a objetos es el de las clases abstractas. Unas clases en las que se pueden definir tanto mtodos como propiedades, pero que <b>no pueden ser instancias directamente.</b> Solamente se pueden usar para construir subclases. Permitiendo as tener una nica implementacin de los mtodos compartidos, evitando la duplicacin de cdigo. Veamos cmo definir las clases abstractas en Python.

## Propiedades de las clases abstractas
<b>La primera propiedad de las clases abstractas es que no puede ser instanciadas.</b> Simplemente proporciona una interfaz para las subclases derivadas y evitando as la duplicacin de cdigo.

Otra caracterstica de estas clases es que no es necesario que tengan una implementacin de todos los mtodos necesarios. Pudiendo ser estos abstractos. Los mtodos abstractos son aquellos que solamente tienen una declaracin, pero no una implementacin detallada de las funcionalidades.

Las clases derivadas de las clases abstractas debe implementar necesariamente todos los mtodos abstractos para poder crear una clase que se ajuste a la interfaz definida. En el caso de que no se defina alguno de los mtodos no se podr crear la clase.

Resumiendo, las clases abstractas define una interfaz comn para las subclases. Proporciona atributos y mtodos comunes para todas las subclases evitando as la necesidad de duplicar cdigo. Imponiendo adems lo mtodos que deber ser implementados para evitar inconsistencias entre las subclases.

## Creacin de clases abstractas en Python
Para poder crear clases abstractas en Python es necesario importar la clase ABC y el decorador abstractmethod del modulo abc (Abstract Base Classes). Un mdulo que se encuentra en la librera estndar del lenguaje, por lo que no es necesario instalar. As para definir una clase privada solamente se tiene que crear una clase heredada de ABC con un mtodo abstracto.

```python
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def mover(self):
        pass
```

Ahora si se intenta crear una instancia de la clase animal, Python no lo permitir indicando que no es posible. Es importante notar que que si la clase no hereda de ABC o contiene por lo menos un mtodo abstracto, Python permitir instancias las clases.

```python
class Animal(ABC):
    #No contiene mtodos abstractos
    def mover(self):
        pass
    
#Permite instacias
Animal()
```

```python
#No hereda de ABC
class Animal():
    @abstractmethod
    def mover(self):
        pass

    #Permite instacias    
Animal()
```

## Mtodos en las subclases
Las subclases tienen que implementar todos los mtodos abstractos, en el caso de que falta alguno de ellos Python no permitir instancias tampoco la clase hija.

```python
class Animal(ABC):
    @abstractmethod
    def mover(self):
        pass
    
    @abstractmethod
    def comer(self):
        print('Animal come')
        
class Gato(Animal):
    def mover(self):
        print('Mover gato')
        
g = Gato() # Error
```

Por otro lado, desde los mtodos de las subclases podemos llamar a las implementaciones de la clase abstracta con el comando super() seguido del nombre del mtodo.

```python
class Animal(ABC):
    @abstractmethod
    def mover(self):
        pass
    
    @abstractmethod
    def comer(self):
        print('Animal come')
        
class Gato(Animal):
    def mover(self):
        print('Mover gato')
             
    def comer(self):
        #Observar que imprime "Animal come" 
        # al llamar al mtodo abstracto "comer"
        super().comer()
        print('Gato come')
        
g = Gato()
g.mover()
g.comer()
```

## <u>Encapsulamiento</u>
El encapsulamiento se considera una de las caractersticas definitorias de la POO.

Cuando una clase existe (se define), se crean objetos a partir de ella, y se usan dichos objetos llamando los mtodos necesarios. Es decir, crea objetos para usar los servicios que nos proporciona la clase a travs de sus mtodos.

No necesita saber cmo trabaja el objeto, ni saber las variables que usa, ni el cdigo que contiene.

El objeto es una caja negra. Es decir, el objeto es un proveedor que proporciona servicios a los clientes que lo solicitan.

La encapsulacin describe el hecho de que los objetos se usan como cajas negras. As, un objeto encapsula datos y mtodos, que estn dentro del objeto.

<b>Interfaz pblica de una clase:</b> Es el conjunto de mtodos (mtodos de servicio) que sirve para que los objetos de una clase proporcionen sus servicios. Estos servicios son los que pueden ser llamados por un cliente.

<b>Mtodos de soporte:</b> Son mtodos adicionales en un objeto que no definen un servicio utilizable por un cliente, pero que ayudan a otros mtodos en sus tareas.

La encapsulacin es un mecanismo de control. El estado (el conjunto de propiedades, atributos  datos) de un objeto slo debe ser modificado por medio de los mtodos del propio objeto.

La tcnica de encapsulacin, es conocida como <b>ocultacin de datos</b>, le permite que los atributos de un objeto pueden ocultarse (superficialmente) para que no sean accedidos desde fuera de la definicin de una clase. Para ello, es necesario nombrar los atributos con un prefijo de doble subrayado: __atributo.

```python
#Crea clase "Factura"
class Factura:
    #Atributo de tipo privado, est oculto, encapsulado 
    __tasa = 19 #Atributo privado
    def __init__(self, unidad, precio):
        self.unidad = unidad
        self.precio = precio
    def por_pagar(self):
        total = self.unidad * self.precio
        #Observar que __tasa se puede usar dentro de la clase propietaria.
        impuesto = total * Factura.__tasa / 100
        return(total + impuesto)

#Instancia objeto "compra1" de tipo "Factura"
compra1 = Factura(12, 110)
#Llamada a atributo "unidad"
print(compra1.unidad)
```

```python
#Llamada a atributo "precio"
print(compra1.precio)
```

```python
#Llamada a mtodo "por_pagar()"
print(compra1.por_pagar(), "bitcoins")
```

```python
#Error al llamar al atributo privado "__tasa"
print(Factura.__tasa)
```

## <u>Polimorfismo</u>

La tcnica de polimorfismo de la POO significa la capacidad de tomar ms de una forma. Una operacin puede presentar diferentes comportamientos en diferentes instancias. El comportamiento depende de los tipos de datos utilizados en la operacin. El polimorfismo es ampliamente utilizado en la aplicacin de la herencia.


```python
class Persona():
    def __init__(self):
        self.cedula = 13765890
    def mensaje(self):
        print("mensaje desde la clase Persona")

class Obrero(Persona):
    def __init__(self):
        self.__especialista = 1
        
    def mensaje(self):
        print("mensaje desde la clase Obrero")

    #def mensaje(self, mensaje):
    #    print(mensaje)

obrero_planta = Obrero()
obrero_planta.mensaje()
#obrero_planta.mensaje("Pepe")
```

Lo que se logra definiendo el mtodo mensaje() en la Clase Derivada (Obrero) se conoce como Mtodo Overriding (cuando se cree el objeto (en este caso obrero_planta y se llame al mtodo mensaje(), este ser tomado de la propia clase y no de la Clase Base Persona). Si comenta o borra el mtodo mensaje() de la clase Obrero (Clase Derivada) y corre nuevamente el cdigo, el mtodo llamado ser el mensaje() de la Clase Base Persona.

```python
class Persona():
    def __init__(self):
        self.cedula = 13765890
    def mensaje(self):
        print("mensaje desde la clase Persona")

class Obrero(Persona):
    def __init__(self):
        self.__especialista = 1
        
    def mensaje(self):
        print("mensaje desde la clase Obrero")

    def mensaje(self, mensaje):
        print(mensaje)

obrero_planta = Obrero()
#obrero_planta.mensaje()
obrero_planta.mensaje("Pepe")
```

## Sobrecarga de Operadores
La sobrecarga de operadores es tambin es conocida por <b>Overloading Operators</b>, trata bsicamente de lo mismo que la sobrecarga de mtodos pero pertenece en esencia al mbito de los operadores aritmticos, binarios, de comparacin y lgicos.

## Tabla de operadores:
En la arquitectura de Python cada objeto tiene varios mtodos especializados que se usan para interactuar con otros objetos o con operadores nativos de Python, regularmente estos mtodos siguen una nomenclatura predefinida en el modelo de datos de Python, esta nomenclatura es la siguiente:

Dado que en Python hay varias funciones nativas tambin se pueden sobrecargar y operar con instancias de objetos, a continuacin, se muestra una tabla con las funciones que se pueden sobrecargar.

```python
class Vector(object):

    def __init__ (self, x, y):
        self.x = x
        self.y = y

    def __add__ (self, v):
        # Sumar dos Vectores
        return Vector (self.x + v.x, self.y + v.y)

    def __sub__ (self, v):
        # Restar dos Vectores
        return Vector (self.x - v.x, self.y - v.y)

    def __mul__ (self, s):
        # Multiplicar un Vectores por un escalar
        return Vector (self.x * s, self.y * s)

    def __repr__ (self):
        # <__main__.Vector instance at 0x01DDDDC8>.
        return '<Vector (%f, %f)>' % (self.x, self.y)
```

```python
#Suma vectorial sobrecargada
a = Vector (3, 5)
b = Vector (2, 7)
print(a + b) # Output: <Vector (5.000000, 12.000000)>
```

```python
#Resta vectorial sobrecargada
a = Vector (3, 5)
b = Vector (2, 7)
c = b - a
print(c) # Output: <Vector (-1.000000, 2.000000)>
```

```python
#Producto vectorial sobrecargada
a = Vector (3, 5)
b = Vector (2, 7)
print(b * 1.3) # Output: <Vector (2.600000, 9.100000)>
```
